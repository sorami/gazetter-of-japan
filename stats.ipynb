# データの確認

```python
import pandas as pd
```

```python
df = pd.read_json("./gazetteer-of-japan.json")
df
```

## 分類ごとのカウント

```python
df.groupby("class")[["kanji"]].count().sort_values("kanji", ascending=False)
```

## 長い名前

```python
df.reindex(df.kanji.str.len().sort_values(ascending=False).index).head(30)
```

## 重複する名前

```python
pd.set_option('display.max_rows', 500)
```

```python
df[df.duplicated("kanji", keep=False)].sort_values("kanji")
```

## 重複する読み

```python
df[df.duplicated("kana", keep=False)].sort_values("kana")
```
